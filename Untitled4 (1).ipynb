```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#understanding the dataset
df = pd.read_csv("/content/drive/MyDrive/hotel_bookings - hotel_bookings.csv")
print(df.shape)
df.info()
df.describe()
print(df.columns.tolist())
df.head(5)
```

```python
sns.heatmap(df.isnull())
plt.show()
#The country category is missing afew
#Most of the company category is missing and the agent is not missing alot
```

```python
#The solution:
df["country"] = df["country"].fillna("other")
df["company"]=df["company"].fillna(0)
df["children"]=df["children"].fillna(df["children"].mode())
df = df.dropna()
sns.heatmap(df.isnull())
plt.show()
```

```python
#understand the unique values in every column
for column in df.columns:
  print(column, df[column].value_counts())
```

```python
#outliers in numerical columns
numerical_columns = df.select_dtypes(include=["int64","float64"]).columns
plt.figure(figsize=(14,len(numerical_columns)*3))
for index, figure in enumerate(numerical_columns,1):
  plt.subplot(len(numerical_columns),2,index)
  sns.boxplot(y=df[figure])
  plt.title(figure)
plt.tight_layout()
plt.show()
```

```python
#remove the outliers
df = df.query("lead_time>200 and stays_in_weekend_nights < 5.0 and stays_in_week_nights <10 and adults<=5 and children==0 and babies==0 and is_repeated_guest==0 and previous_cancellations==0 and previous_bookings_not_canceled==0 and booking_changes==0 and agent<=250 and days_in_waiting_list==0 and adr<=65 and required_car_parking_spaces==0 and total_of_special_requests<=2")
print(df.head(10))
#dropping the columns with only one value
columns = df.columns.tolist()
for column in columns:
  if len(df[column].unique()) == 1:
    df = df.drop(column,axis=1)
```

```python
#remove duplicates
df = df.drop_duplicates()
print(df.columns.tolist())
#In overall, there were some problems with the outliers and the short range of some columns. In addition to missing a lot of data in multiple rows.
```

```python
#handling string columns
print(df.dtypes)
for column in df.select_dtypes(include=["object"]).columns:
  print(f"column: \n {df[column].value_counts()}")
```

```python
#drop the columns reservation_status and reservation_status_date to avoid data leakage
df= df.drop("reservation_status_date",axis=1)
df= df.drop("reservation_status",axis=1)
#use one hot encoding
other_columns = []
for column in df.select_dtypes(include=["object"]).columns:
  if len(df[column].unique()) < 5:
    name_dummies = pd.get_dummies(df[column],prefix=f"{column} : ")
    df = df.drop(column,axis=1)
    df = pd.concat([df,name_dummies],axis=1)
  else:
    other_columns.append(column)
print(df.columns.tolist())
```

```python
#encoding the other columns
for column in other_columns:
  freq = df[column].value_counts()
  df[column] = df[column].map(freq)
df.head(10)
print(df.shape)
```

```python
#generate new features
#children and babies are removed because most of the rows have zero babies or children
#df["total_guests"] = df['adults'] + 0 + 0 #This property has no importance
df['total_nights'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
#df['is_family'] = df['children'] > 0 or df['babies'] >0 #This property has no importance
df.head(10)
```

```python
#understanding the relationship between features
print(df.corr()>0.3)
```

```python
#final preparation
from sklearn.model_selection import train_test_split
train, test = train_test_split(df,random_state=42,test_size=0.2)
train.to_csv("train.csv")
test.to_csv("test_csv")
```
